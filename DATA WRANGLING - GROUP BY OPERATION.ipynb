1.Drop records where the imdb_score column has missing values (NaN)


```python

titles_df = titles_df.dropna(subset=["imdb_score"])
titles_df
```

2.For each TV show or movie which has NaN value in the age certification column, replace it to be No certification


```python
titles_df['age_certification'] = titles_df['age_certification'].fillna("No certification")
titles_df['age_certification'] 
```

For each TV show or movie which has NaN value in the seasons column, replace it to be the most occured value in the season

```python
#Find the most common value in the 'seasons' column
most_common_season = titles_df['seasons'].mode()[0]
most_common_season
```

```python
#Fill NaN values in the 'seasons' column with the most common value
titles_df['seasons'] = titles_df['seasons'].fillna(most_common_season)
titles_df['seasons']
```

Count the number of movies or TV shows for each age certification.

```python
certification_counts = titles_df.groupby("age_certification").size().reset_index(name="count")
certification_counts 
```

Count the number of movies and TV shows (seperately) produced in each release year

```python
count_by_release_year = titles_df.groupby(["release_year", "type"]).size().reset_index(name="count")
count_by_release_year
```

Calculate the average runtime and imdb score of movies and TV shows for each release year.

```python
average_duration_imdb_score = titles_df.groupby("release_year")[["runtime", "imdb_score"]].mean().reset_index()
average_duration_imdb_score 
```

Calculate the sum of IMDb votes for each genre and find the average TMDB score

```python
#Create a new column genre by exploding the genres column
import ast

new_titles_df = titles_df.copy()
new_titles_df['genres'] = new_titles_df['genres'].apply(ast.literal_eval)
new_titles_df['genres'] 
```

```python
#Explode the genres column
exploded_df = new_titles_df.explode('genres')
exploded_df
```

```python
#Group the DataFrame by the genres column and apply different aggregation functions
genres_votes_scores = exploded_df.groupby('genres').agg({
    'imdb_votes': 'sum',    # Calculate the sum of IMDb votes for each genre
    'tmdb_score': 'mean'    # Calculate the average TMDB score for each genre
}).reset_index()
genres_votes_scores 
```

Calculate the standardized score for TMDB popularity for each movie or TV show within its respective genre (use custom defined function)

```python
#Define a custom function to calculate the standardized score
def standardize_score(x):
    return (x - x.mean()) / x.std()
```

```python
#Group by genres and calculate the standardized score for TMDB popularity for each movie or TV show
titles_df['standardized_tmdb_popularity'] = titles_df.groupby('genres')['tmdb_popularity'].transform(standardize_score)
titles_df['standardized_tmdb_popularity'] 
```

Calculate the average length of titles (number of characters) for each genre (use custom defined function)

```python
#Create a function that calculates the average length of titles

def calculate_average_length(titles):
    lengths = titles.apply(lambda x: len(x) if isinstance(x, str) else 0)
    return lengths.mean()
```

```python
#Create new copy of the dataframe

new_titles_df = titles_df.copy()
new_titles_df
```

```python
#Convert the string of genres to an array of countries

import ast

new_titles_df['genres'] = new_titles_df['genres'].apply(ast.literal_eval)
new_titles_df['genres'] 
```

```python
#Explode the genres column

exploded_df = new_titles_df.explode('genres')
exploded_df
```

```python
#Group by genres and calculate the average length of titles

genre_average_length = exploded_df.groupby('genres')['title'].apply(calculate_average_length).reset_index()
genre_average_length 
```
